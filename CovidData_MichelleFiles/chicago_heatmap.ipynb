```python
#start for loop
#truncating text from zip_code_location column to pull out lat and long and make two new columns 

locs = chicago["zip_code_location"]

for i in range(len(locs)):
    text = chicago["zip_code_location"][i]
    loca = text.index('[')+1
    loca2 = text.index(']')

    truncated_text = text[loca:loca2]
    loca3 = truncated_text.index(',')

    lat = truncated_text[:loca3]
    long = truncated_text[loca3+2:]
    
    chicago.loc[i, 'lat'] = lat
    chicago.loc[i, 'long'] = long

    chicago['lat'] = chicago['lat'].astype(float)
    chicago['long'] = chicago['long'].astype(float)
    
chicago
```

```python
chicago = chicago.sort_values(["lat", "long", "cases_cumulative"], ascending=False)
chicago2 = chicago.groupby(["lat", "long"])["cases_cumulative"].max().reset_index()
chicago2
```

```python
chicago2.to_csv("chi_heatmap.csv", index = False)
```

```python
# Store latitude and longitude in locations
locations = chicago2[["long", "lat"]]
cases = chicago2["cases_cumulative"]
```
