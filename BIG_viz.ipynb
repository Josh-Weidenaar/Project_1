```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepaths in variables
big_key_stats = "big_key_stats.xlsx"
big_by_rev = "big_by_rev.xlsx"
big_by_ind = "big_industry.xlsx"
big_by_race = "big_race.xlsx"
```

```python
# Read data file with pandas library
big_key_stats_df = pd.read_excel(big_key_stats)
big_by_rev_df = pd.read_excel(big_by_rev)
big_by_ind_df = pd.read_excel(big_by_ind)
big_by_race_df = pd.read_excel(big_by_race)
```

```python
# Bar chart for BIG recipients by industry
ind = ["Dining", "Fitness", "Events", "Rec", "Support", "Tourism", "Other"]
amt_ind = big_by_ind_df["Grant Amount (Millions)"]
x_axis_ind = np.arange(len(amt_ind))

# Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis_ind]
plt.xticks(tick_locations, ind)

# Set the limits of the y axis
plt.ylim(0, max(amt_ind)+10)

# Give the chart a title and y label
plt.title("BIG Recipients by Industry")
plt.ylabel("Grants Received (Millions)")

plt.bar(x_axis_ind, amt_ind, color="b", align="center")

# Save an image of the chart and print it to the screen
plt.savefig("../Project_1/BarInd.png")
plt.show()
```

```python
# Pie chart of BIG recipients by industry

# Labels for the sections of our pie chart
ind = ["Dining", "Fitness", "Events", "Rec", "Event Support", "Tourism", "Other"]

# The values of each section of the pie chart
amt_ind = big_by_ind_df["Grant Amount (Millions)"]

# The colors of each section of the pie chart
colors = ["red", "orange", "lightcoral", "lightskyblue", "blue", "purple", "green"]

# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

# Creates the pie chart based upon the values above 
# Automatically finds the percentages of each part of the pie chart autopct="%1.1f%%"
plt.pie(amt_ind, labels=ind, colors=colors, startangle=140)

# Save an image of the chart and print it to the screen
plt.savefig("../Project_1/PieInd.png")
plt.show()
```
